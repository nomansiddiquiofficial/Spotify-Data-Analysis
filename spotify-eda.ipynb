**Set index column from the dataset**

```python
dftracks.set_index('release_date', inplace = True)
dftracks.index = pd.to_datetime(dftracks.index)
dftracks.head()
```

Check Artist on row 18

```python
dftracks.iloc[18,5]
```

**Add a column Duration & convert duration from milisecond to seconds then drop milisecond column**

```python
dftracks["duration"] = dftracks["duration_ms"].apply(lambda x: round(x/1000))
dftracks.drop("duration_ms",inplace = True ,axis = 1)
```

**Take just 4% of the data** 

```python
sample_df = dftracks.sample(int(0.004*len(dftracks)))
```

```python
print(len(sample_df))
```

**Draw Regression Plot b/w Loudness and Energy**

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

**Draw Regression Plot b/w Popularity and Acousticness**


```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness")
```

```python
dftracks['dates'] = dftracks.index.get_level_values('release_date')
dftracks.dates = pd.to_datetime(dftracks.dates)
years = dftracks.dates.dt.year
```

**Visualize the total no. of Songs by Distribution Plot by each year since 1922**

```python
sns.displot(years,discrete = True, aspect= 2, height = 5, kind = "hist").set(title = "Number of songs per year")
```

**Visualize the Duration of songs over the years**

```python
total_dr = dftracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr , ax = ax, errwidth = False).set(title= "Year vs Duration")
plt.xticks(rotation =90 )
```

**To see the avg. duration of Songs over the years by line plot**

```python
total_dr = dftracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr , ax = ax).set(title= "Year vs Duration")
plt.xticks(rotation =60 )
```

```python
df_genre = pd.read_csv("/kaggle/input/ultimate-spotify-tracks-db/SpotifyFeatures.csv")
```

```python
df_genre.head()
```

**Duration of the Songs in Different Genres**

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y= "genre", x= 'duration_ms', data = df_genre)
plt.xlabel("durtaion in milli seconds")
plt.ylabel("Genres")
```

**Top 5 Genres by Popularity**

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize= (10,5))
famous = df_genre.sort_values("popularity",ascending = False).head(10)
sns.barplot(y= "genre", x = 'popularity', data = famous).set(title= "Top 5 Genres by Popularity")
```
